```python
import heapq
import time
```

```python
#taking input of the graph in the form of a dictionary(adjacency list form)
n = int(input('Enter the number of edges : '))
from collections import defaultdict
graph = defaultdict(list)
input_str = """A B 2
A C 5
B C -4
C D 1
C E -2
D E -3
E F 3
F G 2
G H -1
H I 1
I J 4
D G -1
C H 1
B G 2
"""
input_lines = input_str.split('\n')
for i in range(n):
    x,y,z = input_lines[i].split()
    z = int(z)
    graph[x].append((y,z))

start_state = input()
goal_state = input()
```

```python
class Node:
    def __init__(self, state, cost, parent=None):
        self.state = state
        self.cost = cost
        self.parent = parent

    def __lt__(self, other):
        return self.cost < other.cost
```

```python
def ucs(start_state, goal_state, graph):
    #steps variable will take a count of how many steps take place while the excution of the algorithm
    global steps
    steps =0
    #open is the open data structure where nodes are brought to be closed
    open = []
    #closed is the closed data structure where teh closed nodes sets are stored
    global closed
    closed = set()
    #vatiable to keep the track of the maximum memory that is used during the algorithm
    global maxmem
    maxmem =0

    #initialising the algo by pushing th start node
    start_node = Node(start_state, 0)
    heapq.heappush(open, start_node)


    while open:
        current_node = heapq.heappop(open)

        #termination with success
        if current_node.state == goal_state:
            path = []
            global mincost
            mincost = current_node.cost
            while current_node:
                path.append(current_node.state)
                current_node = current_node.parent

            return path[::-1]

        #putting the node in open
        closed.add(current_node)

        #exploring the neighbours
        for neighbor, cost in graph[current_node.state]:
            new_cost = current_node.cost + cost
            flag = 0
            for i in closed:
                if i == neighbor and i.cost > new_cost:
                    i.cost = new_cost
                    heapq.heappush(open, i)
                    closed.remove(i)
                    flag = 1
                    break
            if flag == 1:
                continue
            for i in open:
                if i == neighbor and i.cost > new_cost:
                    i.cost = new_cost
                    flag = 1
                    break
            if flag == 1:
                continue
            else:
                new_cost = current_node.cost + cost
                new_node = Node(neighbor, new_cost, current_node)
                heapq.heappush(open, new_node)
                steps+=1
        maxmem = max(maxmem, len(open)+len(closed))

    #termination with failure - the open is empty and goal state has nit been found
    return None
```

```python
start_time = time.time_ns()
path = ucs(start_state, goal_state, graph)
end_time = time.time_ns()

if path:
    print(f"UCS Algorithm Path from {start_state} to {goal_state}: {path}")
    print(f"The minimum cost is {mincost}")
    print(f"Max memory used(maximum number of nodes stored) = {maxmem}")
    print(f"Time elapsed = {end_time - start_time}")
    print(f"Number of nodes closed = {len(closed)}")
    print(f"Number of steps = {steps}")
else:
    print(f"No path found from {start_state} to {goal_state}")
```
