```python
start_x, start_y = np.where(maze == -1)
start = (start_x[0], start_y[0])
parent_start = (-1, -1)
```

```python
def generate_path(end, parents):
    if end == (-1, -1):
        return "Not Found"
    else:
        path = [end]
        parent = parents[end]
        while parent != (-1, -1):
            path.insert(0, parent)
            parent = parents[parent]
        return path
```

```python
def find_neighbors(node, open, close, parents, pos = 0):
    x, y = node
    if x > 0:
        if (x - 1, y) not in close and (x - 1, y) not in open:
            if maze[x - 1][y] != 2:
                open.insert(pos, (x - 1, y))
                parents[(x - 1, y)] = (x, y)

    if y > 0:
        if (x, y - 1) not in close and (x, y - 1) not in open:
            if maze[x][y - 1] != 2:
                open.insert(pos, (x, y - 1))
                parents[(x, y - 1)] = (x, y)

    if x < size[0] - 1:
        if (x + 1, y) not in close and (x + 1, y) not in open:
            if maze[x + 1][y] != 2:
                open.insert(pos, (x + 1, y))
                parents[(x + 1, y)] = (x, y)

    if y < size[1] - 1:
        if (x, y + 1) not in close and (x, y + 1) not in open:
            if maze[x][y + 1] != 2:
                open.insert(pos, (x, y + 1))
                parents[(x, y + 1)] = (x, y)
```

```python
def dfs(start, maze):
    dfs_steps = 0
    dfs_max_space = 0
    open = [start]
    close = []
    parents = {start: parent_start}
    end = (-1, -1)

    while len(open) > 0:
        dfs_max_space = max(dfs_max_space, len(open) + len(close))
        node = open.pop(0)
        dfs_steps += 1
        if maze[node[0]][node[1]] == 1:
            end = node
            break
        close.append(node)
        find_neighbors(node, open, close, parents)
    return generate_path(end, parents), dfs_steps, dfs_max_space
```

```python
def bfs(start, maze):
    bfs_steps = 0
    bfs_max_space = 0
    open = [start]
    close = []
    parents = {start: parent_start}
    end = (-1, -1)

    while len(open) > 0:
        bfs_max_space = max(bfs_max_space, len(open) + len(close))
        node = open.pop(0)
        bfs_steps += 1
        if maze[node[0]][node[1]] == 1:
            end = node
            break
        close.append(node)
        find_neighbors(node, open, close, parents, pos = -1)
    return generate_path(end, parents), bfs_steps, bfs_max_space
```

```python
def print_path(path):
    print("Path Length:", len(path))
    for i in range(len(maze)):
        for j in range(len(maze[0])):
            if (i, j) in path:
                soln_maze[i][j] = '@'
```

```python
mapping = {'S': '*', ' ': ' ', 'E': '*', 'W': '!'}
soln_maze = np.vectorize(mapping.get)(maze_raw)
```

```python
path, dfs_steps, dfs_max_space = dfs(start, maze)
print_path(path)
print("Nodes Visited:", dfs_steps)
print("Maximum number of Nodes stored:", dfs_max_space)
print(soln_maze)
print("DFS will be preferable when space comlpexity is a major problem or the tree depth is large")
```
